````python
import pyFAI
import pyFAI.test.utilstest
import fabio
from matplotlib.pyplot import subplots
from pyFAI.gui import jupyter
from pyFAI.gui.jupyter.calib import Calibration
import flatdict
import pandas as pd
````

````python
cd /data/id13/inhouse3/DATAPOLICY_I3_1/eh2/inhouse/ihsc1691/id13/XXPROCESS
````

````python
samples = {'holder1':{
                    'pcl-9-50':['roi12181_13982', 
                                'roi12182_13985', 
                                'roi12270_14020'],
                    'pcl-9-100':['roi12174_13962', 
                                 'roi12179_13969', 
                                 'roi12180_13975', 
                                 'roi12268_14017'],
                    'pcl-9-200':['roi12170_13938', 
                                 'roi12171_13948', 
                                 'roi12172_13955', 
                                 'roi12267_14011'],
                    'bkg':['roi12271_14073'],
                    },
           'holder2':{
                   'pcl-7-50':['roi12545_14277',
                               'roi12561_14337'],
                   'pcl-11-50':['roi12540_14268',
                                'roi12541_14271',
                                'roi12542_14274',
                                'roi12543_14264',
                                'roi12556_14305',
                                'roi12559_14316',
                                'roi12560_14326'],
                   'pcl-13-50':['roi12547_14286',
                                'roi12549_14288',
                                'roi12551_14291',
                                'roi12552_14294',
                                'roi12563_14349',
                                'roi12562_14341'],
                   'bkg':['roi12553_14297'],
                    },
           #'calib':[],
          }
````

````python
samples_flat = dict(flatdict.FlatDict(samples, delimiter='/'))

for sample_name in samples_flat.keys():
    for roi in samples_flat[sample_name]:
        roi_name = sample_name+'/'+roi
        
        edf_path = './'+roi_name+'_avg.edf'
        edf_name = roi+'_avg_edf'
        exec('%s = fabio.open(edf_path)' % (edf_name))
        
        dat_path = './'+roi_name+'_avg.dat'
        dat_name = roi+'_avg_dat'
        read_dat_code = '''with open(dat_path) as file: 
            %s=pd.read_table(file, sep='\s+', skiprows = 23, names=['I'])'''
        exec(read_dat_code % (dat_name))
        
        azim_path = './'+roi_name+'_avg.azim'
        azim_name = roi+'_avg_azim'
        exec('%s = fabio.open(azim_path)' % (azim_name))
        
# calib_edf = fabio.open("./XXPROCESS/al2o3_m800.edf")
# ax = jupyter.display(calib_edf.data)
````

````python
fig, ax = subplots()

for sample_name in samples_flat.keys():
    for roi in samples_flat[sample_name]:
        dat_name = roi+'_avg_dat'
        exec ('ax.plot(%s, label=dat_name)' % dat_name)
ax.legend();
````
